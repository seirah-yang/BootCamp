# 1. 자릿수 더하기

image.png

```python
import time
import re

start = time.time()

def solution(n):
# 답안 작성 부분 ===============
    answer = 0
    for i in str(n):
        answer += int(i)
# ==============================
    return answer

print(solution(123))
print(solution(987))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```

# 5. 행렬의 덧셈

image.png

```python
import time
import re

start = time.time()

def solution(arr1,	arr2):
# 답안 작성 부분 ===============
    answer = []
    for i in range(len(arr1)):
        for j in range(len(arr1[i])):
            answer+= [arr1[i][j] +arr2[i][j]]


# ==============================
    return answer

print(solution([[1,2],[2,3]],	[[3,4],[5,6]]))
print(solution([[1],[2]],	[[3],[4]]))

end = time.time()

print("작동시간:", f"{end - start:.5f} sec")
```
